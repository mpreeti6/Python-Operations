```python
n = int(input())
star = 1
space = n-1

for i in range(2*n+1):
    if i<n:
        for j in range(space):
            print("", end = " ")
        for j in range(star):
            print("*", end = " ")
        space = space-1
        star = star + 1
        print()
    elif i==n:
        space = space + 1 
        star = star - 1
    else:
        space = space + 1 
        star = star - 1
        for j in range(space):
            print("", end = " ")
        for j in range(star):
            print("*", end = " ")
        
        print()
```
